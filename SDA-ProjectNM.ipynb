## Prepare the data 

In this step, I will prepare the data for the subsequent analysis. 
This includes calculating the number of calls, texts and volume used per user per month as well as the total monthly revenue per user. I will make separate pivot tables for each variable and then finally merge them into one big dataset containing each user's monthly consumption. Calls and internet used will be rounded up to the next full unit.  Moreover, I will convert calls, messages and internet to the integer type.  

For 10 users, there is no data available on any of the services used. For them, the monthly base revenue rate will be considered. 

```python
calls['call_date']=  pd.to_datetime(calls['call_date'], format='%Y-%m-%d')
calls['month']=  calls['call_date'].dt.month
calls['duration']= calls['duration'].apply(mt.ceil)

calls_pivot= calls.pivot_table(index= ['user_id', 'month'], values= 'duration', aggfunc= ['count', 'sum'])


calls_pivot.columns= ['number_of_calls', 'minutes_used']


calls_pivot.reset_index(inplace= True)

calls_pivot
```

```python
messages['message_date']=  pd.to_datetime(messages['message_date'], format='%Y-%m-%d')
messages['month']=  messages['message_date'].dt.month

message_pivot= messages.pivot_table(index= ['user_id', 'month'], values= 'id', aggfunc= 'count')

message_pivot.columns= ['messages_sent']
message_pivot.reset_index(inplace= True)
message_pivot
```

```python
internet['session_date']=  pd.to_datetime(internet['session_date'], format='%Y-%m-%d')
internet['month']=  internet['session_date'].dt.month

internet['gb_used']= internet['mb_used']/1024

internet_pivot= internet.pivot_table(index= ['user_id', 'month'], values= 'gb_used', aggfunc= ['sum'])

internet_pivot.columns= ['gb_used']

internet_pivot['gb_used']= internet_pivot['gb_used'].apply(mt.ceil)

internet_pivot.reset_index(inplace= True)
internet_pivot
```

```python
merged= internet_pivot.merge(calls_pivot.merge(message_pivot,how= 'outer', on=['user_id', 'month']), how= 'outer', 
                             on= ['user_id', 'month'])

merged['gb_used']= merged['gb_used'].where(merged['gb_used'].notnull(), 0)
merged['number_of_calls']= merged['number_of_calls'].where(merged['number_of_calls'].notnull(), 0)
merged['minutes_used']= merged['minutes_used'].where(merged['minutes_used'].notnull(), 0)
merged['messages_sent']= merged['messages_sent'].where(merged['messages_sent'].notnull(), 0)

#merged.reset_index(inplace= True)
merged.reset_index(inplace= True, drop= True)

merged.head()
```

```python
def exceed_calls(x,minutes): 
    df= plans.query('plan_name == @x')
    if minutes > df['minutes_included'].max():
        xtra_call= (minutes-df['minutes_included'].max())*df['usd_per_minute'].max()
        return (round(xtra_call, 2))
    return 0

def exceed_messages(x,txts): 
    df= plans.query('plan_name == @x')
    if txts > df['messages_included'].max():
        xtra_txt= (txts-df['messages_included'].max())*df['usd_per_message'].max()
        return (round(xtra_txt, 2))
    return 0


def exceed_volume(x,vol): 
    df= plans.query('plan_name == @x')
    gb= df['mb_per_month_included'].max()/1024
    if vol > gb:
        xtra_vol= (vol-gb)*df['usd_per_gb'].max()
        return (round(xtra_vol))
    return 0
```

## Analysis of the data 

Now, I'm going to compare the consumption habits of the users of both plans by looking at the duration of each users' calls , the number of messages sent and the volume of data used per month. This will be mainly done by plotting histograms and comparing mean, median and standard deviation. In order to see whether the observed differences matter, I will also conduct statistical tests of the differences for each of the variables of interest. 

```python
ultimate= merged.query('plan == "ultimate"')
customersUltimate = ultimate.pivot_table(index= 'user_id', values= ['minutes_used','messages_sent', 'gb_used'])
surf= merged.query('plan == "surf"')
customersSurf = surf.pivot_table(index= 'user_id', values= ['minutes_used','messages_sent', 'gb_used'])


for columnName, columnData in customersUltimate.iteritems():
    for colName, colData in customersSurf.iteritems():
        if columnName == colName:
            columnData.plot(kind= 'hist', title= columnName, bins= 5, density= True, label= 'Ultimate')
            colData.plot(kind= 'hist', title= columnName, bins= 5, density= True, label= 'Surf')
            plt.legend()
            plt.show()
```

```python
results= st.mannwhitneyu(ultimate['gb_used'], surf['gb_used'])

alpha= 0.05 

print('pvalue :', results.pvalue)

if results.pvalue < alpha:
    print("The users of the two plans differ significantly in their usage of internet")
else:
    print("The users of the two plans do not differ signifcantly in their usage of internet")
```

```python
results= st.mannwhitneyu(ultimate['minutes_used'], surf['minutes_used'])

alpha= 0.05 

print('pvalue :', results.pvalue)

if results.pvalue < alpha:
    print("The users of the two plans differ significantly in their usage of calls")
else:
    print("The users of the two plans do not differ significantly in their usage of calls")
```

```python
results= st.mannwhitneyu(ultimate['messages_sent'], surf['messages_sent'])


alpha= 0.05 

print('pvalue :', results.pvalue)

if results.pvalue < alpha:
    print("The users of the two plans differ significantly in their usage of text messages")
else:
    print("The users of the two plans do not differ significantly in their usage of text messages")
```

Using an alpha of 0.05, we cannot reject the null hypothesis that the means of both groups are equal for Internet as well as for the duration of calls per month. This seems to be in line with our intuition. As it seems, however, users from the Ultimate plan do send more text messages per month. Nevertheless, the similarity of both groups should be taken into account for further appropriations to the budget as well as the very design of the prepaid plans.

## Test the Hypotheses

Now, I'm going to test two hypotheses regarding the mean revenue of different consumer groups. The first referes to the two plans and can be read as follows: 

**H1 'The mean monthly revenue per user differs between the Surf and Ultimate plan'**

**H0 'The mean monthly revenue per user does not differ between the Surf and Ultimate plan'**

The second test compares the mean revenue generated from users living in the NY-NJ area and the one from all other users. The hypothesis and corresponding null hypothesis are: 

**H1 'The mean monthly revenue per user differs between the NY-NJ area and the rest of the country'**

**H0 'The mean monthly revenue per user does not differ between the NY-NJ area and the rest of the country'**

Before conducting the tests, one should select the appropriate statistical procedure. A very common methodology is to  check if the data is normally distributed and if the variances of both groups are equal. These parameters dictate which test is most suitable for a given problem. While the code below will show this in more detail, the data is not normal and the variances of the groups were in both cases not equal. Thus, the chosen statistical test was the Mann-Whitney-U test. This test is robust against non-normality and unequal variances by assigning equally spaced ranks to the observations instead of using the real numbers. 

####  The mean monthly revenue per user differs between the Surf and Ultimate plan

```python
month_rev['plan']= month_rev['user_id'].map(plan_it)
rev_surf= month_rev.query('plan== "surf"')
rev_ultimate= month_rev.query('plan== "ultimate"')
```

```python
# test for normality 

shapiro_surf= st.shapiro(rev_surf['total_revenue_month'])

shapiro_ultimate =st.shapiro(rev_ultimate['total_revenue_month'])
print(shapiro_surf)

print(shapiro_ultimate)
```

The shapiro test for normality shows that in both cases revenue is not normally distributed (right side of the tuples). 
(Note that I was not able to access the pvalue with the .pvalue method). 

```python
# test for equal variance 

eq_var= st.levene(rev_ultimate['total_revenue_month'], rev_surf['total_revenue_month'])

alpha= 0.05 

print("pvalue:",eq_var.pvalue)

if eq_var.pvalue < alpha:
    print("The variances of both groups are not equal")
else: 
    print("The variances of both groups are equal")
```

The Levene test for variance showed that the variances are unequal in both groups. Thus, a regular t-test would result in incorrect conclusions. For cases like that, the Mann Whitney-U test is the more suitable tool. So let's now check our hypotheses.

```python
mean_surf= rev_surf['total_revenue_month'].mean()
mean_ultimate= rev_ultimate['total_revenue_month'].mean()
median_surf= rev_surf['total_revenue_month'].median()
median_ultimate= rev_ultimate['total_revenue_month'].median()

print('Mean monthly revenue Surf:', mean_surf)
print('Mean monthly revenue Ultimate:', mean_ultimate)
print('Median monthly revenue Surf:', median_surf)
print('Median monthly revenue Ultimate:', median_ultimate)

alpha= 0.05
results= st.mannwhitneyu(rev_surf['total_revenue_month'], rev_ultimate['total_revenue_month'])

print('p_value :', results.pvalue)

if (results.pvalue < alpha):
    print('The mean revenue of both plans is significantly different.')
else:
    print("The mean revenue of both plans is not significantly different." )
```

As it appears, the mean revenue is significantly higher for the Ultimate plan (USD 71.77 vs. USD 57.51). Be that as it may, in terms of mean the baseline revenue difference of USD 50 shrinked to only USD 13.66. When accounting for outliers by looking at median, the difference is larger (USD 24.97) but still much lower than the baseline difference. 
This should definitely be looked at more closely as it implies that slight adjustments to the exceed payment rates of the Surf plan may make the mean revenue difference vanish almost entirely. 

#### The mean monthly revenue per user does not differ between the NY-NJ area and the rest of the country

```python
def region(x):
    for value in range(len(users)):
        if x == users['user_id'][value]:
            return (users['city'][value])
        
month_rev['region']= month_rev['user_id'].map(region)

month_rev.head()
```

```python
rev_NY= month_rev.query('region == "New York-Newark-Jersey City, NY-NJ-PA MSA"')
rev_notNY= month_rev.query('region != "New York-Newark-Jersey City, NY-NJ-PA MSA"')

#test for normality 

shapiro_revNY= st.shapiro(rev_NY["total_revenue_month"])
print(shapiro_revNY)
shapiro_revnotNY= st.shapiro(rev_notNY["total_revenue_month"])
print(shapiro_revnotNY)
```

```python
#equal variances

eq_var2= st.levene(rev_NY['total_revenue_month'], rev_notNY['total_revenue_month'])

alpha= 0.05 

print("pvalue:", eq_var2.pvalue)

if eq_var2.pvalue < alpha:
    print("The variances of both groups are unequal")
else:
    print("The variances of both groups are equal")
```

As we can see from the Shapiro test, the distributions of both groups are not equal, thus ruling out the possibility for a regular t-test. The variances of both groups are, on the other hand, very similar. However, the violation of normality is enough reason to resort to the Mann-Whitney-U test for this test as well. 

```python
mean_NY= rev_NY['total_revenue_month'].mean()
mean_notNY= rev_notNY['total_revenue_month'].mean()
median_NY= rev_NY['total_revenue_month'].median()
median_notNY= rev_notNY['total_revenue_month'].median()

print('Mean revenue from users in the NY-NJ area :', mean_NY)
print('Mean revenue from all other users :', mean_notNY)
print('Median revenue from users in the NY-NJ area :', median_NY)
print('Median revenue from all other users :', median_notNY)

alpha= 0.05 

results= st.mannwhitneyu(rev_NY['total_revenue_month'], rev_notNY['total_revenue_month'])

print('p-value :', results.pvalue)

if (results.pvalue < alpha):
    print('The mean monthly revenue per user does  differ between the NY-NJ area and the rest of the country')
else:
    print('The mean monthly revenue per user does not differ between the NY-NJ area and the rest of the country')
```

Judging from the mean there is no significant difference between the two groups. However, the medians are farther apart from each other so we might want to use a Kruskal-Wallis test which uses the median as a benchmark to conclude if samples are drawn from the same population (see: https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/)

```python
mean_NY= rev_NY['total_revenue_month'].mean()
mean_notNY= rev_notNY['total_revenue_month'].mean()
median_NY= rev_NY['total_revenue_month'].median()
median_notNY= rev_notNY['total_revenue_month'].median()

print('Mean revenue from users in the NY-NJ area :', mean_NY)
print('Mean revenue from all other users :', mean_notNY)
print('Median revenue from users in the NY-NJ area :', median_NY)
print('Median revenue from all other users :', median_notNY)

alpha= 0.05 

results= st.kruskal(rev_NY['total_revenue_month'], rev_notNY['total_revenue_month'])

print('p-value :', results.pvalue)

if (results.pvalue < alpha):
    print('The median monthly revenue per user does  differ between the NY-NJ area and the rest of the country')
else:
    print('The median monthly revenue per user does not differ between the NY-NJ area and the rest of the country')
```

Apparently, users from the NY-NJ area do not spend significantly less than the rest of the users even though the means and medians may at first glance suggest otherwise. While the tests conducted were two-tailed, the result would not change if we halved the p-values to make the hypothesis directed. The key takeaway from this exercise is that not too many resources should be invested in changing NY-NJ users' behaviour.

**5. Recommendations**

While it's clear that Ultimate users on average bring in more revenue than Surf users, one should not blindly conclude that increasing Ultimate's advertising budget will make the business more profitable. After all, judging from the sample, Surf users constitute roughly 2/3 of all clients. This is very likely due to the much lower entry-barrier of the plan. Nevetheless, effectively, they do not spend much less than Ultimate users. So one way to look at the data is that Surf users are in fact the cash cows of the business. Two potential approaches are thinkable from my perspective: 

1. Focus on attracting more customers to the Surf plan and increase payment rates after exceeding the package limits (eg. 5-8 cts for calls and txts, USD 12-15 for internet). That way, Surf users may end up paying the same amount as Ultimate users or directly choose Ultimate. As we are talking about prepaid plans, customers are much less likely to keep track of their actual monthly spending in contrast to a postpaid contract. Moreover, I believe it is much more cost-efficient to persuade people to use the Surf plan. Thus, chances are that this strategy will reap benefits.  

2. Provide incentives to get more people into the Ultimate plan as spending does not differ so much. This could come in the form of generous welcome bonusses.
